# Data Visualization

```python
# Libraries used for Analysis and visualization
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
import sys
sys.path.append('../src')
```

## Visualization Analysis

```python
df = pd.read_excel('../data/raw/GSAF5.xls')
df.head()
```

## Location and State Analysis

```python
# imported custom cleaning function
from utils import clean_location_and_state 
df = clean_location_and_state(df)
```

```python
top_states = df['state_clean'].value_counts().nlargest(6).reset_index()
top_states.columns = ['state_clean', 'Count']
print(top_states)
```

```python

from utils import clean_location_and_state
fig = px.bar(top_states,
             y='state_clean',
             x='Count',
             color='Count',
             color_continuous_scale='Reds',
             title =  'Top 5 States with Most Shark Attacks', 
             text = 'Count')
fig.update_traces(textposition='outside')  # move labels above bars
fig.update_layout(uniformtext_minsize=8, uniformtext_mode='hide')
fig.show()
```
